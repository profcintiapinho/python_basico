## Marcadores

## Criar rtulos para um grfico

Com o Pyplot, voc pode usar as funes xlabel()e ylabel()para definir um rtulo para os eixos x e y.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

## Criar um ttulo

Com o Pyplot, voc pode usar a title()funo para definir um ttulo para o grfico.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

## Definir propriedades de fonte para ttulo e rtulos

Voc pode usar o fontdictparmetro em xlabel(), ylabel()e title()para definir as propriedades da fonte para o ttulo e os rtulos.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## Como usar o ColorMap

Voc pode especificar o mapa de cores com o argumento de palavra-chave cmapcom o valor do mapa de cores, neste caso, 'viridis'que  um dos mapas de cores integrados disponveis no Matplotlib.

Alm disso, voc deve criar um array com valores (de 0 a 100), um valor para cada ponto no grfico de disperso:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

# Grfico de Barras

Com o Pyplot, voc pode usar a bar()funo para desenhar grficos de barras:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

Se voc quiser que as barras sejam exibidas horizontalmente em vez de verticalmente, use a barh()funo:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

## Cor da barra
O bar()and barh()usa o argumento de palavra-chave colorpara definir a cor das barras:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

Ou voc pode usar valores de cor hexadecimais :

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50")
plt.show()
```

# Grficos de pizza Matplotlib

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

Por padro, a plotagem da primeira fatia comea no eixo x e se move no sentido anti-horrio.

## Rtulos
Adicione rtulos ao grfico de pizza com o labelparmetro.

O labelparmetro deve ser uma matriz com um rtulo para cada fatia:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

## Explodir
Talvez voc queira que uma das cunhas se destaque? O explodeparmetro permite que voc faa isso.

O explodeparmetro, se especificado, e no None, deve ser uma matriz com um valor para cada fatia.

Cada valor representa a que distncia do centro cada cunha  exibida:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

## Legenda
Para adicionar uma lista de explicao para cada cunha, use a legend()funo:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/profandersonvanin01/python_introducao/main/housing.csv')
```

```python
dados = dados[['latitude','longitude']]
```

```python
dados.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(dados['longitude'].values,
            dados['latitude'].values,
            s=5, c='blue',
            alpha = 0.5, zorder=5)
```

```python
def plot_map(lat1,lat2,lon1,lon2,xmap,ymap):
  plt.figure(figsize=(xmap,ymap))
  map = Basemap(projection='cyl', resolution='h',
                llcrnrlat=lat1, urcrnrlat=lat2,
                llcrnrlon=lon1, urcrnrlon=lon2)
  map.drawcoastlines()
  map.fillcontinents(color='palegoldenrod', lake_color='lightskyblue')
  map.drawmapboundary(fill_color='lightskyblue')
  map.drawparallels(np.arange(lat1, lat2 + 0.5, 2), labels=[1,0,0,0])
  map.drawmeridians(np.arange(lon1, lon2 + 0.5, 2), labels=[0,0,0,1])
  return map
```
